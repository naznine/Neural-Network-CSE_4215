```python
import numpy as np
import pandas as pd
import math
import itertools

# Data Preparation
no_bit = 5  # Number of input nodes (features)
ds = ["".join(p) for p in itertools.product("01", repeat=no_bit)]  # Generate binary strings
reshaped_ds = np.array(ds).reshape(-1, no_bit)  # Reshape to match (32, 5)

df = pd.DataFrame(reshaped_ds, columns=[f"X{i+1}" for i in range(no_bit)])  # Create DataFrame with column names

# Set desired output for each sample
df["Output"] = df["X1"].astype(int)

# Neural Network Configuration
nodes_in_hidden_layer = 2
class_no = len(set(df["Output"]))

# Randomly initialize weights and thresholds
Wij = np.random.rand(no_bit, nodes_in_hidden_layer)
Uhj = np.random.rand(nodes_in_hidden_layer)
Wjk = np.random.rand(nodes_in_hidden_layer, class_no)
Uok = np.random.rand(class_no)

# Sigmoid Activation Function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Compute Output of Hidden Layer
def computeOutputHidden(row):
    netj = np.dot(row, Wij) + Uhj
    Oj = sigmoid(netj)
    return Oj

# Example Calculation
input_row = np.array([0, 0, 1, 0, 0])
hidden_output = computeOutputHidden(input_row)
print(hidden_output)
```
